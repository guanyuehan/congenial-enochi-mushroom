```python
import sqlite3
from sqlite3 import Error
```

```python
# clear all data from the table
def clear_posts_table(db_file):
    conn = sqlite3.connect(db_file)
    try:
        cursor = conn.cursor()
        cursor.execute("DELETE FROM posts")
        conn.commit()
        print("Posts table cleared successfully.")
    except Error as e:
        print(f"Error clearing posts table: {e}")
    conn.close()

clear_posts_table("database.db")
```
